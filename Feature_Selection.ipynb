```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/Users/ashutoshsharma/Desktop/Python/datasets/titanic.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.drop(['name','cabin','boat','body','home.dest'],inplace=True,axis=1)
```

```python
df.age.fillna(df.age.median(),inplace=True)
df.fare.fillna(df.fare.median(),inplace=True)
# df.cabin.fillna(df.cabin.mode(),inplace=True)
df.embarked.fillna(df.embarked.mode(),inplace=True)
```

```python
df
```

```python
df[['sex','embarked_q','embarked_s']] = pd.get_dummies(df[['sex','embarked']],drop_first=True)
```

```python
df.drop('ticket',axis=1,inplace=True)
```

```python
column = list(df)
```

```python
column.sort()
column.append('survived')

column
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
order_rank = SelectKBest(chi2,k='all')
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X.drop('ticket',axis=1,inplace=True)
```

```python
order_feature=order_rank.fit(X,y)
```

```python
# df[df['ticket']== 'C 4001']
```

```python
df1 = pd.DataFrame(order_feature.scores_,columns = ['Score'])
df2 = pd.DataFrame(X.columns,columns=['Features'])
```

```python
df_final = pd.concat([df1,df2],axis=1)
```

```python
df_final.value_counts().sort_values()
```

```python
df_final.nlargest(10,'Score')
```
